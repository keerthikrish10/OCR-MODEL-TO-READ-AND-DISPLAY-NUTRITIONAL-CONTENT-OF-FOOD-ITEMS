```python
from PIL import Image, ImageEnhance, ImageFilter
import pytesseract
import os
from xml.etree import ElementTree as ET
import labelme
from labelme import utils
import json
```

OCR MODEL USING LABELME LIBRARY TO DISPLAY NUTRITIONAL VALUES OF FOOD ITEMS FROM ITS IMAGE:

```python

def extract_text_from_region(image, region):
    # Extract text from a specified region of the image
    xmin, ymin, xmax, ymax = region
    cropped_image = image.crop((xmin, ymin, xmax, ymax))
    text = pytesseract.image_to_string(cropped_image, lang='eng')
    return text

def process_labelme_annotation(annotation_path):
    tree = ET.parse(annotation_path)
    root = tree.getroot()

    image_path = os.path.splitext(annotation_path)[0] + ".jpg"
    img = Image.open(image_path)

    print(f"Image: {image_path}")
    print("Annotations:")
    for obj in root.findall('.//object'):
        label = obj.find('name').text
        bndbox = obj.find('bndbox')
        xmin = int(bndbox.find('xmin').text)
        ymin = int(bndbox.find('ymin').text)
        xmax = int(bndbox.find('xmax').text)
        ymax = int(bndbox.find('ymax').text)
        
        print(f"Label: {label}, Bounding Box: ({xmin}, {ymin}, {xmax}, {ymax})")

        # Extract and display text inside the bounding box
        text_in_box = extract_text_from_region(img, (xmin, ymin, xmax, ymax))
        print(f"Text inside Bounding Box: {text_in_box}")

        # Save the output as an XML file
        output_data = {
            'label': label,
            'text_in_box': text_in_box,
            'bounding_box': {'xmin': xmin, 'ymin': ymin, 'xmax': xmax, 'ymax': ymax}
        }
        output_xml_path = os.path.splitext(annotation_path)[0] + "_output.xml"
        with open(output_xml_path, 'w') as output_file:
            json.dump(output_data, output_file, indent=2)

        print(f"Output saved to: {output_xml_path}")

if __name__ == "__main__":
    images_folder = r'C:\Users\KEERTHI KRISHANA\OneDrive\Documents\nut_value_dataset'

    for filename in os.listdir(images_folder):
        if filename.endswith(".xml"):
            annotation_path = os.path.join(images_folder, filename)
            print(f"Processing annotation: {annotation_path}")

            try:
                process_labelme_annotation(annotation_path)

            except Exception as e:
                print(f"Error processing {annotation_path}: {e}")
```
